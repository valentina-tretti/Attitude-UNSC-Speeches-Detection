# Checking if lexicoder results are contained in annotated text attitude

## Requirements

```python
import pandas as pd
```

## Loading data

```python
# Loading Annotated Dataset
annotated_data = pd.read_excel('speeches_87.xlsx')
annotated_data
```

```python
# Loading Lexikoder Results
lexikoder_results = pd.read_excel('updated_lexicoder_output_sents.xlsx')
lexikoder_results
```

## Merging both dataframes

```python
merged_df = pd.merge(annotated_data, lexikoder_results, on="filename", how="inner")
merged_df
```

```python
# Dropping unnecessary columns
merged_df = merged_df.drop(['label', 'text'], axis=1)
```

```python
merged_df
```

```python
# Correcting format of 'lexicoder_words_polarity' content
#merged_df['lexicoder_words_polarity'] = merged_df['lexicoder_words_polarity'].str.replace(', ', ' ')
#merged_df['lexicoder_words_polarity'] = merged_df['lexicoder_words_polarity'].str.replace(' ', ', ')
#merged_df['lexicoder_words_polarity'] = merged_df['lexicoder_words_polarity'].str.replace(', ,', '')
#merged_df['lexicoder_words_polarity'] = merged_df['lexicoder_words_polarity'].str.replace(' ', '')
#merged_df['lexicoder_words_polarity'] = merged_df['lexicoder_words_polarity'].str.replace(',', ', ')#
merged_df['lexicoder_words_polarity'] = merged_df['lexicoder_words_polarity'].str.replace(',          ', ', ')
merged_df
```

```python
# Converting strings from column 'lexicoder_words_polarity' into list
merged_df['lexicoder_words_polarity']  = merged_df['lexicoder_words_polarity'] .apply(lambda x: x.split(', ') if isinstance(x, str) else [])
```

```python
merged_df
```

```python
merged_df.to_excel('lexikoder_and_annotations_lists.xlsx')
```

## Comparing results

```python
# 1) Compare if strings from column 'lexicoder' are in column 'annotated_text'
merged_df['present'] = merged_df.apply(lambda row: any(str(word).strip() in row['annotated_text'] for word in row['lexicoder_words_polarity']), axis=1)

# 2) Create new column 'present' which says 'True' if at least one string from column 'lexicoder' is in column 'annotated_text'
#    in case it is not= 'False'.
merged_df['present'] = merged_df['present'].astype(str)

# 3) Create new column 'sum' which counts the total numer of strings in column 'lexicoder' that were found in column'annotated_text'
merged_df['sum'] = merged_df.apply(lambda row: sum(1 for word in row['lexicoder_words_polarity'] if isinstance(word, str) and str(word).strip() in row['annotated_text']), axis=1)

# 4) Create new column 'total_words' that contains the total sum of words contained in column  'lexicoder'.
merged_df['total_words'] = merged_df['lexicoder_words_polarity'].apply(lambda x: sum(len(str(word).strip().split(',')) for word in x if isinstance(word, str)))

```

```python
merged_df
```

```python
merged_df.to_excel('lexicoder_VS_annotations.xlsx')
```

```python
# Maping values in present to boolean values
merged_df['present'] = merged_df['present'].map({'True': True, 'False': False})

# Removing all rows containing False under 'present'
merged_df_True = merged_df[merged_df['present']]
```

```python
merged_df_True
```

```python
# Adding a column 'matched' that includes the word or words from 'lexicoder_words_polarity' that were found under column 'annotated_text'
merged_df_True['matched'] = merged_df_True.apply(lambda row: [word.strip() for word in row['lexicoder_words_polarity'] if str(word).strip() in row['annotated_text']], axis=1)
```
